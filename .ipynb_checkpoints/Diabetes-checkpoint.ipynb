```python
other_medicines = ['metformin',
       'repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride',
       'acetohexamide', 'glipizide', 'glyburide', 'tolbutamide',
       'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone',
       'tolazamide', 'glyburide-metformin', 'glipizide-metformin',
       'glimepiride-pioglitazone', 'metformin-rosiglitazone',
       'metformin-pioglitazone']
```

```python
df_with_med.shape[0]
```

%%time
for index, row in df_with_med.iterrows():
    for col in other_medicines:
        if row[col] != 'No':
            df_with_med['with_med'] = 0
            break;
    

```python
df = df_with_med.loc[:,other_medicines]
```

```python
%%time
with_medication_lst = []
for index_label, row_series in df.iterrows():
    l =row_series.values
    if ['Steady'] in l:
        with_medication_lst.append(1)
    else:
         with_medication_lst.append(0)
        
```

```python
df['with_medication'] = with_medication_lst
```
